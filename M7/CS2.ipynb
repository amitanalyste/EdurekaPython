### 2. a. Attempting the categorical encoding using pandas

```python
dummyHorse = horseDf.copy()
```

```python
for column in categorical_columns:
    print(f"++++++{column}+++++++++")
    print(dummyHorse[column].value_counts())
```

```python
dummyHorseT = pd.get_dummies(dummyHorse, prefix = categorical_columns)
dummyHorse.columns
```
